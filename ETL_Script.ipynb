```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
#import data
income_df = pd.read_csv("Connecticut_Earned_Income_Tax_Credit_By_Town.csv", low_memory=False)
sales_df = pd.read_csv("Real_Estate_Sales_2011-2020_GL.csv", low_memory=False)
print(income_df)
print(sales_df)
```

```python
#DimProperty dataframe

#create dataframe
DimProperty = pd.DataFrame(sales_df.filter(['Serial_Number','Property_Town','Address','Property_Type','Residential_Type']))
#slow changing dimensions loading deltas
DimProperty['Effective_from_date'] = sales_df['Date_Recorded']
DimProperty['Effective_to_date'] = np.nan
DimProperty['is_Current'] = '1'

#add surrogate key
DimProperty = DimProperty.rename_axis('DimPropertyID').reset_index()
DimProperty.DimPropertyID = DimProperty.DimPropertyID + 1

display(DimProperty)
```

```python
#DimSales dataframe

#create dataframe
DimSales = pd.DataFrame(sales_df.filter(['Serial_Number', 'Assessed_Value','Sale_Amount','Sales_Ratio']))

#add surrogate key
DimSales = DimSales.rename_axis('DimSalesID').reset_index()
DimSales.DimSalesID = DimSales.DimSalesID + 1

#correct Serial_Number's data type
DimSales['Serial_Number'] = DimSales['Serial_Number'].astype('Int64')

display(DimSales)
```

```python
#DimTown dataframe

#create dataframe
col1 = pd.DataFrame(income_df.filter(['Income_Town']))
col2 = pd.DataFrame(sales_df.filter(['Property_Town']))
DimTown = pd.concat([col1, col2], axis=1)

#add surrogate key
DimTown = DimTown.rename_axis('DimTownID').reset_index()
DimTown.DimTownID = DimTown.DimTownID + 1

display(DimTown)
```

```python
#DimIncome dataframe

#create dataframe
DimIncome = pd.DataFrame(income_df.filter(['Tax_Year', 'Income_Town', 'CT_AGI']))

#add surrogate key
DimIncome = DimIncome.rename_axis('DimIncomeID').reset_index()
DimIncome.DimIncomeID = DimIncome.DimIncomeID + 1

display(DimIncome)
```

```python
#DimDate dataframe

#create dataframe
col = pd.DataFrame(income_df.filter(['Tax_Year']))
col2= pd.DataFrame(sales_df.filter(['Date_Recorded', 'List_Year']))
DimDate = pd.concat([col, col2], axis=1)

#add surrogate key
DimDate = DimDate.rename_axis('DimDateID').reset_index()
DimDate.DimDateID = DimDate.DimDateID + 1

display(DimDate)
```

```python
#extract Year_Recorded and Month_Recorded for hierarchy
DimDate['Year_Recorded'] = pd.DatetimeIndex(DimDate['Date_Recorded']).year
DimDate['Month_Recorded'] = pd.DatetimeIndex(DimDate['Date_Recorded']).month

#convert data type to whole number
DimDate['Year_Recorded'] = DimDate['Year_Recorded'].astype('Int64')
DimDate['Month_Recorded'] = DimDate['Month_Recorded'].astype('Int64')
DimDate['List_Year'] = DimDate['List_Year'].astype('Int64')
DimDate['Tax_Year'] = DimDate['Tax_Year'].astype('Int64')

#convert string to datetime
#DimDate['Date_Recorded'] = pd.to_datetime(DimDate['Date_Recorded'])

display(DimDate)
```

```python
#FactSales dataframe
#create dataframe
sales_cols = pd.DataFrame(DimSales.filter(['DimSalesID','Serial_Number', 'Assessed_Value', 'Sale_Amount']))
date_col = pd.DataFrame(DimDate.filter(['DimDateID']))
town_col = pd.DataFrame(DimTown.filter(['DimTownID']))
income_cols = pd.DataFrame(DimIncome.filter(['DimIncomeID','CT_AGI']))
FactSales = pd.concat([sales_cols, date_col, income_cols, town_col], axis=1)

#correct data display
FactSales['Serial_Number'] = FactSales['Serial_Number'].astype('Int64')
FactSales['DimSalesID'] = FactSales['DimSalesID'].astype('Int64') 
FactSales['DimDateID'] = FactSales['DimDateID'].astype('Int64')
FactSales['DimTownID'] = FactSales['DimTownID'].astype('Int64')
FactSales['DimIncomeID'] = FactSales['DimIncomeID'].astype('Int64')

display(FactSales)
```

```python
#FactIncome dataframe
#create dataframe
date_cols = pd.DataFrame(DimDate.filter(['DimDateID']))
town_cols = pd.DataFrame(DimTown.filter(['DimTownID']))
income_cols = pd.DataFrame(DimIncome.filter(['DimIncomeID']))
cols = pd.DataFrame(income_df.filter(['Tax_Year', 'Income_Town', 'Fed_AGI', 'CT_AGI']))
FactIncome = pd.concat([date_cols, town_cols, income_cols, cols], axis=1, join="inner")

#correct data type display
FactIncome['DimDateID'] = FactIncome['DimDateID'].astype('Int64')
FactIncome['DimTownID'] = FactIncome['DimTownID'].astype('Int64')
FactIncome['DimIncomeID'] = FactIncome['DimIncomeID'].astype('Int64')


display(FactIncome)
```
